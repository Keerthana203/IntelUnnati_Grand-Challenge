```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Alert': ['cas_ldw', 'cas_ldw', 'cas_ldw', 'cas_ldw', 'cas_ldw', 'cas_ldw'],
    'Date': ['01-06-2022', '01-06-2022', '01-06-2022', '01-06-2022', '01-06-2022', '01-06-2022'],
    'Time': ['05:36:36', '05:36:47', '05:37:04', '05:37:31', '05:38:08', '05:38:21'],
    'Lat': [12.887403, 12.892586, 12.894932, 12.898267, 12.901594, 12.903877],
    'Long': [80.083412, 80.085794, 80.087564, 80.090062, 80.092538, 80.094272],
    'Vehicle': [2846, 2846, 2846, 2846, 2846, 2846],
    'Speed': [48, 58, 57, 49, 58, 54]
}

df = pd.DataFrame(data)
```

```python
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df.drop(['Date', 'Time'], axis=1, inplace=True)
```

```python
summary_stats = df['Speed'].describe()
print(summary_stats)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(df['DateTime'], df['Speed'], marker='o', linestyle='-')
plt.title('Vehicle Speed Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Speed (km/h)')
plt.grid(True)
plt.show()
```

```python
#Speed Distribution
plt.figure(figsize=(6, 4))
plt.hist(df['Speed'], bins=10, edgecolor='black')
plt.title('Speed Distribution')
plt.xlabel('Speed (km/h)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Average Speed
average_speed = df.groupby(df['DateTime'].dt.hour)['Speed'].mean()
plt.figure(figsize=(6, 4))
average_speed.plot(kind='bar')
plt.title('Average Speed by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Speed (km/h)')
plt.grid(axis='y')
plt.show()
```

```python
#Location Analysis
plt.figure(figsize=(6, 4))
plt.plot(df['Long'], df['Lat'], marker='o', linestyle='-', color='blue')
plt.title('Vehicle Path')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
#Speed Vs Time
plt.figure(figsize=(10, 6))
plt.scatter(df['DateTime'], df['Speed'], marker='o', color='green', alpha=0.7)
plt.title('Speed vs. Time')
plt.xlabel('Date and Time')
plt.ylabel('Speed (km/h)')
plt.grid(True)
plt.show()
```
